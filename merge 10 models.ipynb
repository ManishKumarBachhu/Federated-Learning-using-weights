```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np

df = pd.read_csv("mnist 60k train data.csv")
```

```python
x = []
y = []
Y = df['Label']
df = df.drop(['Label'], 1)

for i in range(0,10):
    y.append(Y[(i)*6000:(i+1)*6000])
    y[i] = y[i].values
    
for i in range(0,10):
    x.append(df[(i)*6000:(i+1)*6000])
j = 0
for i in range(0,10):
    print(str(j) + ": \n")
    print(x[i].head)
    print(y[i])
    j = j+1
```

```python
from sklearn import preprocessing
mm_scaler = preprocessing.MinMaxScaler()
X_train = []
for i in x:
    X_train.append(mm_scaler.fit_transform(i))
```

```python
def model_build() :
    model1 = tf.keras.models.Sequential()
    model1.add(tf.keras.layers.Dense(20, input_shape = (784,), activation='relu', use_bias = False))
    model1.add(tf.keras.layers.Dense(20, activation='relu', use_bias = False))
    model1.add(tf.keras.layers.Dense(10, activation='softmax'))
    model1.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    
    return model1
```

```python
from sklearn.model_selection import train_test_split as tts
from tensorflow.keras import *

models = []
metrics = [[]]

for l in range(0,10):
    print(str(l) + ": ")
    x_train, x_test, y_train, y_test = tts(X_train[l], y[l], test_size = 0.1, random_state = 0)
    models.append(model_build())
    models[l].fit(x_train, y_train, epochs = 20)
    loss, acc = models[l].evaluate(x_test, y_test)
    metrics.append([loss, acc])
```

```python
sum = 0
metrics2 = metrics[1:]
for i in metrics2:
    sum = sum + i[1]
    print(i)

average = sum/10
print(average)
```

```python
test = pd.read_csv("mnist 10k test data.csv")
x_test = test.drop(['Label'], 1)
x_test = mm_scaler.fit_transform(x_test)
y_test = test['Label']
```

```python
for i in range(0,10):
    print(models[i].evaluate(x_test, y_test))
```

```python
m = models[0].get_weights()
for num in range(1,10):
    a = models[num].get_weights()
    m = np.add(m,a)

model = model_build()
model.set_weights(m)
loss, acc = model.evaluate(x_test, y_test)
print(acc)
print(loss)  
```
