```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import sys

df = pd.read_csv("gender_voice_dataset.csv")
INPUT_SHAPE = 20
LABEL = 'label'
SPLIT_SIZE = 10
```

```python
from sklearn.preprocessing import LabelEncoder
x = df.drop([LABEL], 1)
y = df[LABEL]
print(y)
'''
for i,j in enumerate(y):
    if j == 'male':
        y.loc[i] = 0
    else:
        y.loc[i] = 1
'''
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(y)
y = encoder.transform(y)
print(y)
from sklearn.model_selection import train_test_split
x, x_test, y, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

X_train, Y_train = list(), list()
for i in range(0,SPLIT_SIZE):
    X_train.append(x[int((i*len(x)/SPLIT_SIZE)):(int((i+1)*len(x)/SPLIT_SIZE))])
    Y_train.append(y[(int(i*len(x)/SPLIT_SIZE)):(int((i+1)*len(x)/SPLIT_SIZE))])
    
print(Y_train)
```

```python
def model_build() :
    model1 = tf.keras.models.Sequential()
    model1.add(tf.keras.layers.Dense(20, input_shape = (INPUT_SHAPE,), activation='relu', use_bias = False))
    model1.add(tf.keras.layers.Dense(20, activation='relu', use_bias = False))
    model1.add(tf.keras.layers.Dense(1, activation='sigmoid'))
    model1.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    
    return model1
```

```python
models = list()
metrics = list()
for i in range (0,SPLIT_SIZE):
    models.append(model_build())
    models[i].fit(X_train[i], Y_train[i], epochs = 100)

for i in range(0,SPLIT_SIZE):
    metrics.append(models[i].evaluate(x_test, y_test))
```

```python
sum = 0
metrics2 = metrics
for i in metrics2:
    sum = sum + i[1]

average = sum/SPLIT_SIZE
print(average)
```

```python
m = models[0].get_weights()
for num in range(1,SPLIT_SIZE):
    a = models[num].get_weights()
    m = np.add(m,a)

model = model_build()
model.set_weights(m)
loss, acc = model.evaluate(x_test, y_test)
print(acc)
print(loss)  
```
