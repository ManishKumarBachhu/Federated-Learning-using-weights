```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np

df = pd.read_csv("mnist 60k train data.csv")
```

```python
x1 = df[:30000]
x2 = df[30000:]
print(x1.head, x2.head)
y1 = x1['Label']
y2 = x2['Label']
x1 = x1.drop(['Label'], 1)
x2 = x2.drop(['Label'], 1)
```

```python
from sklearn import preprocessing
mm_scaler = preprocessing.MinMaxScaler()
x1 = mm_scaler.fit_transform(x1)
x2 = mm_scaler.fit_transform(x2)
print(x1, x2)
```

```python
from sklearn.model_selection import train_test_split as tts
x1_train, x1_test, y1_train, y1_test = tts(x1, y1 , test_size = 0.1, random_state = 0)
y1_train = y1_train.values
```

```python
x2_train, x2_test, y2_train, y2_test = tts(x2, y2 , test_size = 0.1, random_state = 0)
y2_train = y2_train.values
```

```python
model1 = tf.keras.models.Sequential()
#model1.add(tf.keras.layers.Flatten())
model1.add(tf.keras.layers.Dense(128, input_shape = (784,), activation='relu'))
model1.add(tf.keras.layers.Dense(128, activation='relu'))
model1.add(tf.keras.layers.Dense(10, activation='softmax'))
model1.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
          
```

```python
model1.fit(x1_train, y1_train, epochs = 5)
loss, acc = model1.evaluate(x1_test, y1_test)
print(acc)
print(loss)
```

```python
model2 = tf.keras.models.Sequential()
#model2.add(tf.keras.layers.Flatten())
model2.add(tf.keras.layers.Dense(128, input_shape = (784,), activation='relu'))
model2.add(tf.keras.layers.Dense(128, activation='relu'))
model2.add(tf.keras.layers.Dense(10, activation='softmax'))
model2.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model2.fit(x2_train, y2_train, epochs = 5)
loss, acc = model2.evaluate(x2_test, y2_test)
print(acc)
print(loss)
```

```python
loss, acc = model2.evaluate(x1_test, y1_test)
print(acc)
print(loss)
```

```python
test = pd.read_csv("mnist 10k test data.csv")
x_test = test.drop(['Label'], 1)
x_test = mm_scaler.fit_transform(x_test)
y_test = test['Label']
```

```python
a1 = model1.get_weights()
print("-----------Model1 weights---------------------")
print(a1)
print()
a2 = model2.get_weights()
print("-----------Model2 weights---------------------")
print(a2)
print()
m = np.add(a1,a2)
print("-----------Add weights---------------------")
print(m)
m /= 2
print("-----------Final weights---------------------")
print(m)
model1.set_weights(m)
print(model1.evaluate(x_test, y_test))
```
